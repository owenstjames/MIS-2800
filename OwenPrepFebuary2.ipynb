```python
import sqlite3
```

```python
conn = sqlite3.connect('NFL_Scouting.db')
```

```python
cursor = conn.cursor()

print("Database connected successfully")
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS scouting_stats (
    team_name TEXT,
    city TEXT,
    wins INTEGER,
    points_per_game REAL
)
""")
conn.commit()
print("Table created successfully")
```

```python
cursor.execute("""
INSERT INTO scouting_stats (team_name, city, wins, points_per_game)
VALUES
    ('Chiefs', 'Kansas City', 11, 29.2),
    ('Bills', 'Buffalo', 10, 26.5),
    ('Cowboys', 'Dallas', 12, 27.8),
    ('Jets', 'New York', 7, 18.6)
""")

conn.commit()
print("Data inserted successfully")
```

```python
cursor.execute("SELECT * FROM scouting_stats WHERE points_per_game > 24")
results = cursor.fetchall()

print("Teams averaging more than 24 points per game:")
for row in results:
    print(row)
```

```python
print("\nStep 5: Teams sorted by wins (highest to lowest):")
cursor.execute("SELECT * FROM scouting_stats ORDER BY wins DESC")
for row in cursor.fetchall():
    print(row)
```

```python
print("\nStep 6: Teams with more than 5 wins, sorted by city:")
cursor.execute("SELECT * FROM scouting_stats WHERE wins > 5 ORDER BY city ASC")
for row in cursor.fetchall():
    print(row)
```

```python
conn.close()
print("\nStep 7: Database connection closed")
```
